```python
import os

import pandas as pd
import pymongo
```

## ENV Variable setup

```python
DB_NAME = os.environ["DB_NAME"]
COLLECTION_NAME = os.environ["COLLECTION_NAME"]
DB_CONNECTION_URL = os.environ["DB_CONNECTION_URL"]

DATASET = "heart_attack_prediction_dataset.csv"
```

```python
df = pd.read_csv(DATASET)
df.head()
```

## MongDB setup

```python
client = pymongo.MongoClient(DB_CONNECTION_URL)
data_base = client[DB_NAME]
collection = data_base[COLLECTION_NAME]
```

Uploading data to mongodb

```python
# df should be converted into dict before we push it to mongodb
data = df.to_dict(orient="records")
record = collection.insert_many(data)
```

getting data from monogodb

```python
df = pd.DataFrame(list(collection.find()))
df.head()
```
